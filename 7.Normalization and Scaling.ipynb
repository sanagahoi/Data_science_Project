## 1.  Data Preparation for scaling

### 1.1 Encode categorical features

```python
## create a copy of the dataframe
df1 = df.copy()
```

```python
df_final = pd.get_dummies(df1, columns= ['Type','Content Rating'], drop_first = True)
df_final.head()
```

### 1.2 Drop irrelevant columns depending the objective

```python
dff_final = df_final.drop(['App','Price'], axis = 1)
```

```python
dff_final.head()
```

### 1.3 Min-Max Scaling (Normalization)

```python
# Scale the numerical features ratings and reviews using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# create an  object
min_max = MinMaxScaler()

df_scaled1 = dff_final.copy()

df_scaled1.iloc[:,:2] = min_max.fit_transform(df_scaled1.iloc[:,:2])
df_scaled1.head()
```

### 1.4 Max absolute scaling

```python
# Scale only numerical features reviews and ratings using MaxAbsScaler
from sklearn.preprocessing import MaxAbsScaler
mas = MaxAbsScaler()# create an  object

df_scaled2 = dff_final.copy()

df_scaled2.iloc[:,:2] = mas.fit_transform(df_scaled2.iloc[:,:2])
df_scaled2.head()
```

### 1.5 Standard Scaler (Standardization)

```python
# Scale only numerical features reviews and ratings using StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()# create an  object

df_scaled3 = dff_final.copy()

df_scaled3.iloc[:,:2] = sc.fit_transform(df_scaled3.iloc[:,:2])
df_scaled3.head()
```
