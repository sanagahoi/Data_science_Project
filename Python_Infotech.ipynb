Write a python function flatten_list(nested) that takes a nested list of arbitrary depth an dreturns a flat list of all elements.

```python
def flatten_list(nested):
    result = []
    for item in nested:
        if isinstance(item, list):
             result.extend(flatten_list(item))
        else:
            result.append(item)
    return result
```

```python
flatten_list([2,3,[4,5,[5,6]]])
```

```python
nested_list=[2,3,[4,5,[5,6]]]
```

```python
def flatten_generator(nested):
    for item in nested:
        if isinstance(item, list):
            yield from flatten_generator(item)
        else:
            yield item

# To get a list back:
flat_list = list(flatten_generator(nested_list))
```

```python
import itertools

nested_shallow = [[2, 3], [4, 5], [5, 6]]
flat_list = list(itertools.chain.from_iterable(nested_shallow))
flat_list
```
