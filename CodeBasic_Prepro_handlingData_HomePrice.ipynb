```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'area':[2600,3000,3200,3600,4000],
    'price':[550000, 570000,600000, 650000, 700000],
    'bedroom':[3,4,None,3,5],
    'age':[20,15,18,30,5]
})
df
```

```python
df.to_csv('homeprice.csv')
```

```python
df=pd.read_csv('homeprice.csv')
df
```

```python
dumm = pd.get_dummies(df.town)
```

```python
dumm
```

```python
merged = pd.concat([df, dumm], axis='columns')
```

```python
merged
```

```python
final = merged.drop(columns=['town', 'west windsor'], axis=1) #drop one town columns to avoid multicolinearity
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X=final.drop('price',axis=1)
```

```python
y=final.price
```

```python
model.fit(X,y)
```

```python
# robinsville home
model.predict([[2800,0,1]])
```

```python
# monroe township home
model.predict([[2800,1,0]])
```

```python
# west windsor home
model.predict([[2800,0,0]])
```

```python
model.score(X,y)
```

```python
# One hot encoding - label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('homeprice.csv')
df
```

```python
le = LabelEncoder()
le_town = le.fit_transform(df.town)
le_town
```

```python
df1=df
```

```python
df1.town =le_town
```

```python
df1
```

```python
X=df1[['town','area']].values
y =df1.price
```

```python
X,y
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False)
```

```python
encoded = ohe.fit_transform(X[:,[0]])
encoded
```

```python
import numpy as np
area_col = X[:,[1]]
X=np.hstack((encoded,area_col))
X
```

```python
X=X[:,1:]
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[0,2600]])
```
