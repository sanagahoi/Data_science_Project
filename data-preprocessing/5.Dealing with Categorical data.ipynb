## Dealing with categorical features - Type and Content Rating

```python
# create a copy of the dataframe
df1 = df.copy()
```

```python
df1['Type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
df1['Content Rating'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

## Encode categorical features

#### Lable Encoding

```python
# Label encoding 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Label encode the categorical feature 'Content Rating '
df_content_rating_le = pd.DataFrame(le.fit_transform(df1['Content Rating']),columns=['Content_rating_LE'])
df_content_rating_le.head()
```

```python
df_content_rating_le['Content_rating_LE'].unique()
```

```python
df1['Content Rating'].unique()
```

#### One hot encoding

```python
## One hot encoding categorical features 'Type' and 'Content Rating'
df_final = pd.get_dummies(df1, columns= ['Type','Content Rating'], drop_first = True)
df_final.head()
```

#### Manual Encoding

```python
### Manual encode the categorical feature 'Type'
df1['Type'].unique()
```

```python
## dropping nan entries before encoding
df1.dropna(inplace=True)
```

```python
df1['Type'].unique()
```

```python
df1['Type'] = df1['Type'].replace({'Free':0,'Paid':1})
```
