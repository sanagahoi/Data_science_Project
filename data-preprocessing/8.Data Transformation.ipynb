## Data Transformation

Check summary statistics of column 'reviews'

```python
## create a copy of the dataframe
df1 = df.copy()
```

```python
df1['Reviews'].describe()
```

Check the distribution of the reviews using boxplot and kdeplot

```python
# reviews distibution 
g = sns.kdeplot(df1['Reviews'], color="Blue", shade = True)
plt.title('Distribution of Reviews',size = 20)
plt.show()
```

```python
sns.boxplot(df1['Reviews'])
plt.show()
```

- We can observe that the reviews data is skewed towards right.

## Log Transformation 

##### Lets apply log transformation on the column reviews and reduce the skewness

```python
df_log = df1.copy()
df_log['Reviews']  = np.log(df_log['Reviews'])
df_log.head()
```

```python
sns.kdeplot(df_log['Reviews'], color="blue", shade = True)
plt.show()
```

```python
sns.boxplot(df_log['Reviews'])
plt.show()
```

- We can observe from above plots after applying log transformation on the column reviews the skewness is been reduced significantly and the distribution is approximately equal to normal distribution
