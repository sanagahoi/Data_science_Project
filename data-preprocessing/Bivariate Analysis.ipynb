## 1. Statistical methods
### 1.1 Heatmap

```python
# creates a correlation matrix of data
corr_matrix = df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr() 
corr_matrix
```

```python
# heatmap helps in visualizing correlation amongst features
sns.heatmap(corr_matrix,annot=True,cmap='plasma',vmin=-1,vmax=1)
plt.show()
```

- Number of colonoies is having high positive correlation with the columns total production, stocks and production value.
- Yield per colony is having negative correlation with the column priceperlb.

## 2. Continous Vs Continous
### 2.1 Scatterplot

#### Let us understand the relationship between the variables 'numcol' and 'prodval' through scatterplot

```python
sns.scatterplot(df['numcol'],df['prodvalue'])
plt.show()
```

- We can observe from the above scatter plot between 'number of colonies' and 'production value', there is strong positive relationship between the variables
- As the number of colonies increases there is a significant increase in the production value

## 3. Continous Vs Categorical
### 3.1 Boxplot

#### Let us understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot

```python
plt.figure(figsize=(10,5))
sns.boxplot(df['year'],df['prodvalue'])
plt.show()
```

- As we can observe from the above plot, the production value got increased year by year.
- We can see that the production value is lesser in the years 1998 to 2001 and in coming years the production value of the honey got increased significantly


## 4. Categorical Vs Categorical
### 4.1 Crosstab

We will take a sample from the dataset to reduce the no. of categories

```python
## creating a sample dataframe
df_sample = df[df['state'].isin(['AL','CO','SC','NV','MD'])]
```

```python
## Lets do crosstab
pd.crosstab(df_sample['state'],df_sample['year'])
```

- From the above results for the crosstab of two categorical variables 'state' and 'year', what we can observe is
- For the state named "MD" there are no entries(records) from  year '2004' to '2012'
- For the state named "SC" there are no entires from 1998-2000 and 2004 to 2012
- For the state named "NV" the last entry is in the year 2008

### 4.2 Stacked Bar Graph

For plotting stacked bar graph also will use the same data created above "df_sample"

```python
pd.crosstab(df_sample['year'],df_sample['state']).plot(kind='bar',stacked=True)
plt.show()
```

- As we can observe from the Stacked Bar chart
- There are no entires for state "SC" from year 1998-200 and 2004-2012
- And for state "MD  there are no entires from year 2004-2012
- For state 'NV' there are no entires from year 2009-2012
