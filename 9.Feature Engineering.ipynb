## Feature Engineering

#### Let's create a new feature as 'Rating_category' by grouping ratings to high and low categories(>3.5 is high rest low)

```python
## create a copy of the dataset
df1 = df.copy()
```

```python
# Method-1: convert ratings to high and low categories using loops
Rating_category = dict()
for i in range(0,len(df1['Rating'])):
    if df1['Rating'].iloc[i] >= 3.5:
        Rating_category[i] = 'High'
    else: Rating_category[i] = 'Low'
        
#Add the categorical column to the data 
final_df = df1.assign(Rating_category = pd.Series(Rating_category).values)
final_df.head(5)
```

```python
# Method-2: using lambda function
df1['category'] = df1.apply(lambda x:'high' if (x['Rating'])>3.5 else 'low', axis=1)
df1.head(5)
```

Check count of high and low ratings

```python
## lets check the percentage of records in each category
df1['category'].value_counts(normalize=True)*100
```

```python
df1['category'].value_counts().plot(kind='bar')
plt.show()
```

- Rating category is highly imbalanced.
