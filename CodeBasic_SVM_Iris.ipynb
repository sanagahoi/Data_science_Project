```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
 df['target'] =iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal Length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='*')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='gray')
```

```python
plt.xlabel('sepal Length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='gray', marker="+")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
X.head()
```

```python
y=df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model = SVC(C=10)  #c is regularization
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model = SVC(gamma=10)  #c is regularization
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model = SVC(kernel='linear')  #c is regularization
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
