Scale Features before apply PCA.
Accuracy might drop.
PCA is called dimensionality reduction technique as it can help us reduce dimensions.

```python
from sklearn.datasets import load_digits
import pandas as pd
```

```python
dataset = load_digits()
dataset.keys()
```

```python
dataset.target
```

```python
import numpy as np
np.unique(dataset.target)
```

```python
df =pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
X=df
y =dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =train_test_split(X_scaled, y , test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
```

```python
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_  #first point - 14% variance it got
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca=train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)
```

It's not improved

```python
pca = PCA(n_components=2)
X_pca =pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca=train_test_split(X_pca, y, test_size=0.2, random_state=30)
model = LogisticRegression(max_iter=10000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)
```
