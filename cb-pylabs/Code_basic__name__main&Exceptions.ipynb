```python
# Multi inheritance
class Father:
    def skills(self):
        print("Programming and gardening")
        
class Mother:
    def skills(self):
        print("Documentation and cooking")
        
class Child(Father, Mother):
    def skills(self):
        Father.skills(self)
        Mother.skills(self)
        print("Singing,  sports")
        
c = Child()
c.skills()
```

```python
# Remote control system

class RemoteControl():
    def __init__(self):
        self.channels = ['HBO', 'CNN','ABC','ESPN']
        self.index=-1
        
    def __iter__(self):
        return self
    def __next__(self):
        self.index+=1
        if self.index == len(self.channels):
            raise StopIteration
        return self.channels[self.index]
    
    
rm = RemoteControl()
itr = iter(rm)
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
```

```python
# fibbonaci series

def fibo():
    a, b=0,1
    while True:
        yield a
        a,b = b, a+b
        
for f in fibo():
    if f>50:
        break
    print(f, end=" ")
    
    
```

```python
# Decorators allows us to make code more readable
import time
def time_count(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(func.__name__+'took : '+str((end-start)*1000)+'ms')
    return wrapper
    
@time_count
def calc_square(nums):
    result = []
    for n in nums:
        result.append(n*n)
    return result

@time_count
def cal_cube(nums):
    result=[]
    for n in nums:
        result.append(n*n*n)
    return result

arr = range(1, 10000)
out_sqr = calc_square(arr)
out_cube = cal_cube(arr)
    
```

```python
# multi threading - juggles between multiple tasks

import time
def calc_square(nums):
    for n in nums:
        time.sleep(0.2)
        print('square',n*n )


def cal_cube(nums):
    for n in nums:
        time.sleep(0.2)
        print('cube',n*n*n)

arr = range(1, 10)
t = time.time()
calc_square(arr)

cal_cube(arr)
print((time.time()-t)*1000, 'ms')

# when we use time.sleep() at that time multithreading tries to utilize that idle time
# so let's use multithreading
```

```python
import time
import threading

def calc_square(nums):
    for n in nums:
        time.sleep(0.2)
        print('square',n*n )


def cal_cube(nums):
    for n in nums:
        time.sleep(0.2)
        print('cube',n*n*n)

arr = range(1, 10)
t = time.time()
th1 = threading.Thread(target=calc_square, args=(arr,))
th2 = threading.Thread(target=cal_cube, args=(arr,))

th1.start()
th2.start()

th1.join()
th2.join()


print((time.time()-t)*1000, 'ms')
```

#### Multiprocessing

Every process has its own address space (virtual memory). Thus program variables are not shared between two processes. You need to use interprocess communication (IPC) technique if you want to share data between two processes.

```python
import multiprocessing
import time

square_result =[]

def calc_square(nums):
    global square_result
    for n in nums:
        print("square: "+ str(n*n))
        square_result.append(n*n)
    print("within the process: ", square_result)
        
if __name__ =='__main__':
    arr=[2,4,25,6]
    p1 = multiprocessing.Process(target=calc_square, args=(arr,))
    p1.start()
    p1.join()
    print('result'+str(square_result))
    print('-'*10,'done', '-'*10)
```

```python
# Sharing Data Between Processes Using Multiprocessing Queue

import multiprocessing

def calc_square(nums, q):
    for n in nums:
        q.put(n*n)
if __name__ == '__main__':
    nums = [2,3,4]
    q = multiprocessing.Queue()
    p = multiprocessing.Process(target=calc_square, args=(nums, q))
    p.start()
    p.join()
    
    while q.empty() is False:
        print(q.get())
```

```python
# multiprocessing lock
# when 2 processes try to access the shared memory or databases, it can create problem to secure db we use lock


import time
import multiprocessing

def deposit(balance):
    for i in range(100):
        time.sleep(0.01)
        balance.value = balance.value +1

def withdraw(balance):
    for i in range(100):
        time.sleep(0.01)
        balance.value = balance.value -1
        
if __name__ == '__main__':
    balance = multiprocessing.Value('i', 2000)
    d = multiprocessing.Process(target=deposit, args=(balance,))
    w = multiprocessing.Process(target=withdraw, args=(balance,))
    
    d.start()
    w.start()
    d.join()
    w.join()
    
    print(balance.value) #it will be executing multiple assembly line transaction so the balance value will become inconsistant
```

```python
# multiprocessing lock
# when 2 processes try to access the shared memory or databases, it can create problem to secure db we use lock


import time
import multiprocessing

def deposit(balance, lock):
    for i in range(100):
        time.sleep(0.01)
        lock.acquire()
        balance.value = balance.value +1
        lock.release()

def withdraw(balance):
    for i in range(100):
        time.sleep(0.01)
        lock.acquire()
        balance.value = balance.value -1
        lock.release()
        
if __name__ == '__main__':
    balance = multiprocessing.Value('i', 2000)
    lock = multiprocessing.Lock()
    d = multiprocessing.Process(target=deposit, args=(balance,lock))
    w = multiprocessing.Process(target=withdraw, args=(balance,lock))
    
    d.start()
    w.start()
    d.join()
    w.join()
    
    print(balance.value) 
```

```python
# multiprocessing pool
# works in py file
# it works on map reduce method and reduces time
from multiprocessing import Pool
import time

def foo(n):
    sum = 0
    for x in range(10):
        sum+=x
    return sum

if __name__ == '__main__':
    t1= time.time()
    p = Pool()
    result = p.map(foo, range(10))
    p.close()
    p.join()
    
    print("pool took: ", time.time()-t1)
    
    t2 = time.time()
    result=[]
    for x in range(10):
        result.append(f(x))
        
    print('Serial processing took: ', time.time()-t2)
    
```
