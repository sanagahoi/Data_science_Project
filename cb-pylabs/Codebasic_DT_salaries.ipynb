```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.DataFrame({
    'company': ['google','google','google','google','google','google','google'],
    'job':['sales executive','sales executive', 'business manager','business manager', 'computer Programmer', 'data scientist', 'software engineer'],
    'degree': ['bachelors','masters','bachelors', 'masters','bachelors','masters', 'bachelors']
})
df.to_csv('salaries.csv', index=False)
```

```python
df = pd.read_csv('salaries.csv')
df.head()
```

```python
import numpy as np
series = pd.Series(np.random.randint(0,2,len(df['company'])))
```

```python
df.rename(columns={0:'salary_more_than_100k'},inplace=True)
```

```python
df.to_csv('salaries.csv')
```

```python
# let's start the work now

df.columns
```

```python
inputs = df.drop('salary_more_than_100k', axis='columns')
target = df['salary_more_than_100k']
```

```python
le_company =LabelEncoder()
le_job= LabelEncoder()
le_degree=LabelEncoder()
```

```python
# we can also add columns like this 

inputs['company_name'] =le_company.fit_transform(inputs['company'])
inputs['job_title'] =le_job.fit_transform(inputs['job'])
inputs['degree_level'] =le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'], axis='columns')
inputs_n
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

```python
model.predict([[2,2,1]])
```
