```python
# Create a new file and write it
f = open("file.text", 'w')
f.write("love python")
f.close()
```

```python
# overwrite a file
f = open("file.text", 'w')
f.write("I love Java")
f.close()
```

```python
# To add additional in the same file
f = open('file.text', 'a')
f.write("I love ML")
f.close()
```

```python
# to add data in new line 
f = open('file.text', 'a')
f.write("\nI love DS")
f.close()
```

```python
# to read the lines at a time
f= open('file.text', 'r')
print(f.read())
f.close()
```

```python
# read line by line
f= open('file.text', 'r')
for line in f:
    print(line)
    
f.close()
```

```python
f= open('file.text', 'r')
for line in f:
    tokens = line.split(" ")
    print(tokens)
f.close()
```

```python
f= open('file.text', 'r')
for line in f:
    tokens = line.split(" ")
    print(len(tokens))

f.close()
```

```python
# to open another file keeping wordcount
f=open('file.text', 'r')
f_out= open('word_count.txt','w')
for line in f:
    tokens = line.split(' ')
    f_out.write( " word count : "+ str(len(tokens))+" "+line)

 
f.close()
f_out.close()
```

```python
# to open file witn reading and writing - r+
# to open file witn writing and reading - w+ - it creates file if it doesn't exist

# with used to automatically close the file without explicitly writing it

with open('file.text', 'r') as f:
    print(f.read())
    
print(f.closed)
```
