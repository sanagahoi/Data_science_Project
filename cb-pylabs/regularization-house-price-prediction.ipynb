```python
import pandas as pd
nrowsreed=1000
# The path must have NO space between the dots
file_path = '../input/datasets/anthonypino/melbourne-housing-market/Melbourne_housing_FULL.csv'

# Read the data
df = pd.read_csv(file_path, nrows=nrowsreed)

# Preview the data
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'].fillna(df.BuildingArea.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['YearBuilt','Lattitude','Longtitude'], axis='columns', inplace=True)
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
X = df.drop("Price", axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

It is performing poorly on the test data so let's try using L1 regularization (Lasso regularization) to improve performance

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg
```

```python
# L1 regularization is adding absolute value to the regression model
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

It's visible that Lasso regression is not helping in improving the model. let's try L2 regularization in the same model

```python
from sklearn.linear_model import Ridge
rg = Ridge(alpha=50, max_iter=100, tol=0.1)
rg.fit(X_train, y_train)
```

```python
rg.score(X_train, y_train)
```

```python
rg.score(X_test, y_test)
```
