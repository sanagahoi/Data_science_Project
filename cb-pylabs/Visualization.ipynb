```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [2,4,5,6,5]
y = [34,54,23,44,65]
```

```python
plt.plot(x,y)
plt.xlabel('day')
plt.ylabel("temperature")
plt.title("Weather")
```

```python
plt.xlabel('day')
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

```python
plt.xlabel('day')
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y, 'g*--')
```

```python
plt.xlabel('day')
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y, 'g+--',linestyle='')
```

```python
plt.xlabel('day')
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y, 'rd',linestyle='')
```

```python
plt.xlabel('day')
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y, 'rd',linestyle='', alpha=.5)
```

```python
days=[1,2,3,4,5,6]
max_temp =[51,53,76,56,45,55]
min_temp = [32,23,25,24,33,34]
avg_temp = [45,35,50,37,39,44]
```

```python
plt.plot(days, max_temp, label='max')
plt.plot(days,min_temp, label='min')
plt.plot(days,avg_temp, label='avg')
plt.legend(loc='best', shadow=True)
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('Weather')
plt.grid()
```

```python
profit =[30,23,34,21]
```

```python
plt.xticks(yps,company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Stock Price')
plt.bar(xps,revenue, label='Revenue')
plt.bar(xps,profit, label='Profit')
plt.legend()
```

```python
plt.xticks(yps,company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Stock Price')
plt.bar(xps-0.2,revenue, label='Revenue')
plt.bar(xps+0.2,profit, label='Profit')
plt.legend()
```

```python
plt.xticks(yps,company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Stock Price')
plt.bar(xps-0.2,revenue,width=0.4, label='Revenue')
plt.bar(xps+0.2,profit,width=0.4, label='Profit')
plt.legend()
```

```python
plt.yticks(yps,company)
plt.ylabel('Company')
plt.xlabel('Revenue')
plt.title('US Stock Price')
plt.barh(xps-0.2,revenue, label='Revenue')
plt.barh(xps+0.2,profit, label='Profit')
plt.legend()
```

```python
# HISTOGRAM

blood_sugar_men = np.random.randint(80,200,20)
blood_sugar_men
```

```python
plt.hist(blood_sugar_men)
```

```python
plt.hist(blood_sugar_men, bins=[80,100,125,150])
```

```python
plt.hist(blood_sugar_men, bins=[80,100,125,200],rwidth=0.8)
```

```python
plt.hist(blood_sugar_men, bins=[80,100,125,200], color= 'g')
```

```python
plt.hist(blood_sugar_men, bins=[80,100,125,200], color= 'g',histtype='step')
```

```python
blood_sugar_women = np.random.randint(80,200,20)
```

```python
blood_sugar_women
```

```python
plt.hist([blood_sugar_men, blood_sugar_women])
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,200],rwidth=0.8, label=['men','women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,200],rwidth=0.8, label=['men','women'], orientation='horizontal')
plt.legend()
```

```python
# pie chart

exp_vals =[1400,300,500,600]
exp_lbl = ['Home rent', 'Meds', 'food','essentials']
```

```python
plt.pie(exp_vals, labels=exp_lbl)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_lbl,radius=1,autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_lbl,radius=1,autopct='%0.1f%%',shadow=True, explode=[0,0.1,0,0.1])
plt.show()
```
