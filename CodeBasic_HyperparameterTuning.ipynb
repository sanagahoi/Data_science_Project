```python
from sklearn import svm, datasets
```

```python
iris = datasets.load_iris()
```

```python
import pandas as pd
df =pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['flower'] = iris.target
```

```python
df['flower'] =df['flower'].apply(lambda x:iris.target_names[x])
df[47:52]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size=0.2)
```

```python
model= svm.SVC(kernel='rbf',C=30, gamma='auto')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),iris.data,iris.target, cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10,gamma='auto'),iris.data,iris.target, cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target, cv=5)
```

```python
import numpy as np
kernels =['rbf', 'linear']
C=[1,10,20]
avg_scores={}
for kval in kernels:
    for cval in C:
        cv_scores= cross_val_score(svm.SVC(kernel=kval,C=cval,gamma='auto'),iris.data, iris.target,cv=5)
        avg_scores[kval+'_'+str(cval)] =np.average(cv_scores)
avg_scores
```

```python
# the same thing we can do with single line of code with gridsearchcv

from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C': [1,10,20],
    'kernel':['rbf','linear']
},cv=5, return_train_score=False)

clf.fit(iris.data, iris.target)
```

```python
clf.cv_results_    #not feasible view, no issues we can convert it into df
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

Grid search cv has a costly computation cost bcs it tries permutation and combination at every parameter

```python
# to reduce it we can use randomizedSearchCV bcs it uses random combination of parameters

from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
}, cv=5,
                       return_train_score=False,
                       n_iter=2)
```

```python
rs.fit(iris.data, iris.target)
```

```python
pd.DataFrame(rs.cv_results_)
```

```python
# how to choos and find which model works best for the data
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params ={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rf','linear']
        }
    },
    'random-forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'logistic-regression':{
        'model':LogisticRegression(),
        'params':{
            "C":[1,5,10]
        }
    }
}
```

```python
scores = []
for model, mp in model_params.items():
    clf =GridSearchCV(mp['model'],mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append({
        'model':model,
        'best score': clf.best_score_,
        'best params':clf.best_params_
    })
```

```python
df = pd.DataFrame(scores, columns=
                 ['model', 'best score', 'best params'])
df
```
